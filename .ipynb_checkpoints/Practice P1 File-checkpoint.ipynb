```python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import os
os.listdir("test_images/")
%matplotlib inline
```

```python
import math

def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def canny(img, low_threshold, high_threshold):
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    mask = np.zeros_like(img)   
    
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

```python
# Reading in an image
image = mpimg.imread('test_images/solidYellowCurve2.jpg')
```

```python
kernel_size = 5
low_threshold = 60
high_threshold = 150
ysize = image.shape[0]
xsize = image.shape[1]

x_noise = ((xsize)*(1/13))
y_noise = ((ysize)*(1/14))
vertices = np.array([[(((xsize/2)+x_noise),ysize/2),(xsize/2,((ysize/2)+y_noise)), (xsize, ysize), (0,ysize)]], dtype=np.int32)
rho = 1
theta = np.pi/180
threshold = 10
min_line_len = 12
max_line_gap = 5
line_image = np.copy(image)*0


# Make image grayscale 
gray_img = grayscale(image)

# Average 
gaussian_img = gaussian_blur(gray_img, kernel_size)
canny_edges = canny(gaussian_img, low_threshold, high_threshold)
masked_edges = region_of_interest(canny_edges, vertices)
color_edges = np.dstack((masked_edges, masked_edges, masked_edges)) 

# F
lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]), min_line_len, max_line_gap)
```

```python
plt.imshow(masked_edges)
```

```python
leftSlopes = []
rightSlopes = []
leftLines = []
rightLines= []
```

```python
for line in lines:
    for x1,y1,x2,y2 in line:
        slope = (y2-y1)/(x2-x1)
        if slope > 0.25:
            leftSlopes.append(slope)
            leftLines.append(line)
        elif slope < -0.25:
            rightSlopes.append(slope)
            rightLines.append(line)
```

```python
print(leftSlopes)
print(rightSlopes)
```

```python
leftSlope = np.mean(leftSlopes)
rightSlope = np.mean(rightSlopes)
print('This is the left slope:', leftSlope)
print('This is the right slope:', rightSlope)
```

```python
leftYMin = ysize
leftYMax = 0
leftXs = []
leftYs = []
for line in leftLines:
    for x1,y1,x2,y2 in line:
        leftXs.append(x1)
        leftXs.append(x2)
        leftYs.append(y1)
        leftYs.append(y2)

        if y1 < leftYMin:
            leftYMin = y1
            associatedLXmin = x1
        if y2 < leftYMin:
            leftYMin = y2
            associatedLXmin = x2

        if y1 > leftYMax:
            leftYMax = y1
            associatedLXmax = x1
        if y2 > leftYMax:
            leftYMax = y2
            associatedLXmax = x2

leftMeanX = np.mean(leftXs)
leftMeanY = np.mean(leftYs)
leftB = leftMeanY - (leftSlope*leftMeanX)
```

```python
print("Left Slope:", leftSlope)
print("Left Mean X:", leftMeanX)
print("Left Mean Y:", leftMeanY)
print("Y Intercept:", leftB)
print("Left Y Min:", leftYMin)
print("Left Y Max:", leftYMax)
print(leftXs)
```

```python
rightYMin = ysize
rightYMax = 0
rightXs = []
rightYs = []
associatedXmin = 0
associatedXmax = 0
for line in rightLines:
    for x1,y1,x2,y2 in line:
        rightXs.append(x1)
        rightXs.append(x2)
        rightYs.append(y1)
        rightYs.append(y2)

        if y1 < rightYMin:
            rightYMin = y1
            associatedXmin = x1
        if y2 < rightYMin:
            rightYMin = y2
            associatedXmin = x2

        if y1 > rightYMax:
            rightYMax = y1
            associatedXmax = x1
        if y2 > rightYMax:
            rightYMax = y2
            associatedXmax = x2

rightMeanX = np.mean(rightXs)
rightMeanY = np.mean(rightYs)
rightB = rightMeanY - (rightSlope*rightMeanX)
```

```python
print("Right Slope:", rightSlope)
print("Right Mean X:", rightMeanX)
print("Right Mean Y:", rightMeanY)
print("Y Intercept:", leftB)
print("Right Y Min:", rightYMin)
print("Right Y Max:", rightYMax)
```

```python
y_Max = leftYMax
if rightYMax > y_Max:
    y_Max = rightYMax

y_Min = leftYMin
if rightYMin > y_Min:
    y_Min = rightYMin
    
if leftYMax != y_Max:
    leftYMax = y_Max
    associatedLXmax = int((y_Max-leftB)/leftSlope)
elif rightYMax != y_Max:
    rightYMax = y_Max
    associatedXmax = int((y_Max-rightB)/rightSlope)
    
if leftYMin != y_Min:
    leftYMin = rightYMin
    associatedLXmin = int((leftYMin-leftB)/leftSlope)
elif rightYMin != y_Min:
    rightYMin = leftYMin
    associatedXmin = int((rightYMin-rightB)/rightSlope)
    
    
```

```python
cv2.line(image,(associatedLXmin,leftYMin),(associatedLXmax,leftYMax),(255,0,0),10)
cv2.line(image,(associatedXmin,rightYMin),(associatedXmax, rightYMax),(255,0,0),10)
lines_edges = cv2.addWeighted(color_edges, 0.8, image, 1, 0) 
plt.imshow(image)
```
